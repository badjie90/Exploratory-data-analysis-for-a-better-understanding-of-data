### Let's analyse the missing values in the data

We will check the percentage of missing values present in each feature of the datset

```python
## 1 -step make the list of features which has missing values
features_with_MV=[features for features in house_p.columns if house_p[features].isnull().sum()>1]

## 2- step print the feature name and the percentage of missing values

for feature in features_with_MV:
    print(feature, np.round(house_p[feature].isnull().mean(), 4),  ' % missing values')
```

### Let's see the Numerical Variables

```python
# list of numerical variables
numerical_features = [feature for feature in house_p.columns if house_p[feature].dtypes != 'O']

print('Number of numerical variables: ', len(numerical_features))

# visualise the numerical variables
house_p[numerical_features].head()
```

#### Let's analyse the temporal Variables(Eg: Datetime Variables) in the dataset

#### The Dataset includes four variables related to years. We can obtain details such as the number of years or days from the datetime variables. As an instance, we can compute the gap in years between the construction year of the house and the year it was sold, in this particular situation. This analysis will be carried out in the Feature Engineering section.

```python
# list of variables that contain year information
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]

year_feature
```

```python
# let's explore the content of these year variables
for feature in year_feature:
    print(feature, house_p[feature].unique())
```

#### Lets analyze the Temporal Datetime Variables. We will check whether there is a relation between year the house is sold and the sales price

```python
year_feature
```


#### Let's visualize and compare the difference between All the years feature with SalePrice by using a scatter plot


```python
#### Numerical variables are usually of 2 type
#### 1. Continous variable and Discrete Variables

discrete_feature=[feature for feature in numerical_features if len(house_p[feature].unique())<25 and feature not in year_feature+['Id']]
print("Discrete Variables Count: {}".format(len(discrete_feature)))
```

```python
discrete_feature
```

```python
house_p[discrete_feature].head()
```

```python
continuous_feature=[feature for feature in numerical_features if feature not in discrete_feature+year_feature+['Id']]
print("Continuous feature Count {}".format(len(continuous_feature)))
```

#### Lets analyse the continuous values by creating histograms to visualize and understand the distribution

```python
for feature in continuous_feature:
    data=house_p.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

### Let's check the Outliers in the data. Outliers are considered to be very high or low values in the dataset. They are normally very far from the main distribution of the data.

#### It is important to note that, checking the outliers using a box-plot is only possible for continous features, is not not possible for categorical features

```python
for feature in continuous_feature:
    data=house_p.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
        
    
```
